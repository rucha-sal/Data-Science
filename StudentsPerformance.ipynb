```python
#duplicate
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

duplicates = df[df.duplicated()]

if duplicates.empty:
    print("No duplicate rows found.")
else:
    print("Duplicate rows found:")
    print(duplicates)
```
